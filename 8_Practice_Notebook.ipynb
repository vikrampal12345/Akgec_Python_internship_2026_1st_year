### 1.1 — Your First While Loop (Countdown)


````python
# A while loop runs as long as the condition is True.
# Three ingredients: (1) starting value, (2) condition, (3) update

count = 5            # ① starting value
while count > 0:     # ② condition — checked before every round
    print(count)
    count += 1       # ③ update — keeps us from looping forever

print("Liftoff!")
````

### 1.2 — Counting Up


````python
# Loop that counts up to 10
num = 1
while num <= 10:
    print(num)
    num += 1

print("Done counting.")
````

### 1.4 — break: Exit the Loop Immediately


````python
# Search for the first number divisible by 7 between 1 and 50
num = 1
while num <= 50:
    if num % 7 == 0:
          print(f"Found it: {num}")
          break        # stop searching — we found what we needed
    num += 1
````

### 1.5 — continue: Skip This Round Only


````python
# Print only odd numbers from 1 to 9
num = 0
while num < 10:
    num += 1 #1
    if num % 2 == 0:
      continue
    else:
      print(num)
````

### 1.6 — while True + break (Input Validation)


````python
# Keep asking until the user enters a valid positive number
while True:
    user_input = input("Enter a positive number: ")
    if user_input.isdigit() and int(user_input) > 0:
        number = int(user_input)
        continue
    print("That's not a positive number. Try again.")

print(f"Great — you entered {number}.")
````

### 1.7 — ATM PIN Simulator (Capstone Demo)


````python
# ATM PIN simulator — 3 attempts, then locked out
# Concepts used: while, if, break, and compound condition

correct_pin = 1234
attempts = 3

while attempts > 0:
    guess = int(input("Enter your PIN: "))
    if guess == correct_pin:
        print("✓ Access granted!")
        break
    attempts -= 1
    if attempts > 0:
        print(f"Incorrect PIN. {attempts} attempt(s) remaining.")

if attempts == 0:
    print("✗ Account locked. Please visit your nearest branch.")
````

### Q1 — What are the three ingredients every while loop needs?


````python
# Predict the answer, then run to reveal.
answer_q1 = """
① A starting value — set before the loop begins
② A condition  — checked before each round; loop exits when it becomes False
③ An update    — changes the variable inside the body so the condition eventually fails
"""
print(answer_q1)
````

### Q2 — What is the difference between `break` and `continue`?


````python
answer_q2 = """
break    → exits the entire loop immediately. Execution continues after the loop.
continue → skips the remaining lines of THIS round and jumps back to the condition check.
           The loop itself is not exited.
"""
print(answer_q2)
````
