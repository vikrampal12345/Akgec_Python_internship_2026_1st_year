### 1.1  Creating a list

````python
# A list holds multiple values in one variable
marks = [85, 92, 78, 95, 88]

print(type(marks))     # <class 'list'>
print(len(marks))      # 5
type(print(marks))         # [85, 92, 78, 95, 88]
````

### 1.1b  Indexing and mutability

````python
# Indexing — same as strings from Day 4
print(marks[0])        # 85  (first item)
print(marks[-1])       # 88  (last item)
print(marks[2])        # 78  (third item)

# But unlike strings, you can CHANGE a list item
marks[2] = 99 #mutability
print(marks)           # [85, 92, 99, 95, 88]  — 78 became 99!
````

### 1.2  Slicing (quick review — same rules as strings)

````python
marks = [85, 92, 78, 95, 88]

print(marks[1:4])      # [92, 78, 95]  — stop is excluded
print(marks[:3])       # [85, 92, 78]  — from start
print(marks[::2])      # [85, 78, 88]  — every other item
````

### 1.3  List methods — modify the list in place

````python
marks = [85, 92, 78]
print("Start:", marks)

# .append(val) — add to the end
marks.append(72)
print("After append(72):", marks)

# .insert(i, val) — add at position i
marks.insert(0, 100)
print("After insert(0, 100):", marks)

# .pop() — remove from the end
removed = marks.pop()
print("After pop():", marks, "— removed:", removed)

# .pop(i) — remove at position i
removed2 = marks.pop(2)
print("After pop(2):", marks, "— removed:", removed2)

# .remove(val) — remove first occurrence of a value
marks.remove(85)
print("After remove(85):", marks)
````

````python
# .sort() — sort ascending
marks = [85, 92, 78, 95, 88]
marks.sort()
print("Sorted:", marks)

# .sort(reverse=True) — sort descending
marks.sort(reverse=True)
print("Reversed:", marks)

# ⚠️ COMMON TRAP: .sort() returns None!
marks = [85, 92, 78]
result = marks.sort()
print("result =", result)      # None!
print("marks =", marks)        # [78, 85, 92] — the list itself changed
````

### 1.4  Iterating through a list

````python
marks = [85, 92, 78, 95, 88]

# Basic for loop
for mark in marks:
    print(mark)
````

````python
# Accumulator pattern — you know this from Day 10!
total = 0
for mark in marks:
    total += mark

average = total / len(marks)
print(f"Total: {total}")
print(f"Average: {average}")
````

### 1.5  Tuples — fixed collections

````python
# A tuple uses parentheses () — it's like a list that can't change
point = (3, 7)
student = ("Riya", 19, "CSE")

print(point[0])        # 3  — indexing works
print(student[1])      # 19
````

````python
x = [3,7]
x[0] = 6
print(x)
````

````python
# Tuples are IMMUTABLE — trying to change them crashes
try:
    point[0] = 5
except TypeError as e:
    print(f"Error: {e}")
    print("Tuples cannot be changed after creation!")
````

````python
# Build a marks tracker that collects, sorts, and generate average
all_marks = []
for i in range(5):
  marks = int(input(f"Write marks in subject {i+1}:"))
  all_marks.append(marks)
all_marks.sort(reverse=True)
total = 0
for i in all_marks:
  total+=i
average = total/len(all_marks)
print(f'Your average marks is : {average}')
````

## Section 4 — Self-Check

**Q1:** What does `marks.sort()` return?

<details><summary>Answer</summary>

`None`. The `.sort()` method modifies the list **in place** and returns `None`.
To see the sorted list, print the list itself after calling `.sort()`.

```python
marks.sort()       # sorts the list — returns None
print(marks)       # THIS shows the sorted list
```
</details>

---

**Q2:** Can you change the first element of a tuple?

<details><summary>Answer</summary>

No. Tuples are **immutable** — once created, their values cannot be changed.
`point[0] = 5` raises a `TypeError`. If you need to change values, use a list instead.
</details>

---

**Q3:** What is the difference between `.append()` and `.insert()`?

<details><summary>Answer</summary>

- `.append(val)` adds the value **at the end** of the list
- `.insert(i, val)` adds the value **at position i**, shifting everything else right

```python
items = [10, 20, 30]
items.append(40)       # [10, 20, 30, 40]
items.insert(0, 5)     # [5, 10, 20, 30, 40]
```
</details>
