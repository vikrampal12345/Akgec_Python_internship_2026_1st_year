### 1.1 — The Dot Syntax
A method is a function that *belongs to* a string.
You call it with: `string . method_name ()`


```python
# The dot notation: string . method()
x = 'hello'
x.upper()
```

```python
# Methods return a NEW string — the original doesn't change
name = "hello"
name.upper()        # returns "HELLO" but doesn't change name
```

```python
# To keep the result, store it
name = "hello"
clean = name.upper()
print(clean)        # HELLO
print(name)         # hello  — original unchanged
```

### 1.3 — Cleaning: strip and replace


```python
# .strip() — remove leading and trailing whitespace
messy = " amit mishra "
clean = messy.strip()
print(messy)   # shows the spaces
print(clean)   # spaces gone
```

```python
# .replace(old, new) — swap one substring for another
email = "riya  @gmail.com"
fixed = email.replace("riya", "amit")
fixed = fixed.replace(" ","")
#fixed = email.strip()
print(fixed)         # riya@gmail.com
#print(email[4:].strip())
```

```python
word = 'Chess'
word_new = word.replace("he","she")
print(word_new)
```

```python
name =  'ShivNarayan'
clean = name.replace('S',"A")
clean = clean.replace('A','y')
print(clean)
```

### 1.4 — Searching: find and count


```python
# .find(sub) — index of first match, or -1 if not found
word = "riya"
print(word.find("r"))
print(word.find("ya"))    # 3  (index 3)
print(word.find("z"))    # -1 (not found — this is NOT an error)
```

```python
# .count(sub) — how many times sub appears
sentence = "riya sharma"
print(sentence.count("h"))    # 3
```

### 1.6 — Method Chaining: the Week 1 Capstone
Chain methods left-to-right: the output of one becomes the input of the next.


```python
# Chaining: strip first, then title
messy = "  rIYA sHARMA  "

# Step 1: strip() removes leading/trailing spaces
# Step 2: title() fixes the case on the already-clean string
clean = messy.strip().title()
print(clean)         # Riya Sharma
```

```python
# Try it with your own messy string:
your_name = "  AMIT mishra  "
print(your_name.strip().title())
```

### Stretch A — Username Generator
Build a username from a full name:
