### 1.1 · Arithmetic Operators

```python
# The five standard arithmetic operators
a = 10
b = 3

#round(x,2)

print(a + b)   # Addition 13
print(a - b)   # Subtraction 7
print(a * b)   # Multiplication 30
print(round((a / b),4))   # Division  ← what type does this give? 3.33
print(a ** b)  # Exponentiation (a to the power b) 1000
print(a//b) #3
print(a%b) #1
```

### 1.2 · Floor Division and Modulo

```python
# Floor division: whole-number result (decimal part dropped)
print(17 // 5)   # How many full 5s fit in 17? 3
print(7  // 2)   # How many full 2s fit in 7? 3

# Modulo: the remainder after division
print(17 % 5)    # 17 = (5 × 3) + 2  →  remainder is 2
print(20 % 4)    # 20 = (4 × 5) + 0  →  remainder is 0

# Useful trick: even/odd checker
print(7 % 2)     # 1 means odd
print(8 % 2)     # 0 means even

#prime checker
```

### 1.4 · Comparison Operators
These return `True` or `False` — a new type of value called a **bool**.

```python
x = 10
y = 3

print(x == y)   # Equal to? False
print(x != y)   # Not equal to? True
print(x > y)    # Greater than? True
print(x < y)    # Less than? False
print(x >= 10)  # Greater than or equal? True
print(y <= 3)   # Less than or equal? True

# Check the type
print(type(x > y))  # <class 'bool'>
```

```python
# KEY REMINDER: = stores.  == asks a question.
score = 95        # Stores 95 in score
print(score == 95)  # Asks: is score equal to 95?  → True
print(score == 80)  # Asks: is score equal to 80?  → False
```

### 1.5 · Assignment Shorthand

```python
# Cricket scorer — long form
score = 0
score = score + 4   # boundary
score = score + 6   # six
score = score + 1   # single
print("Long form:", score)

# Same thing — shorthand
score = 0
score += 4   # boundary
score += 6   # six
score += 1   # single
print("Shorthand:", score)  # Same result

# Works for subtraction and multiplication too
health = 100
health -= 20   # took damage
print("Health:", health)

price = 50
price *= 2     # doubled
print("Price:", price)
```

### 1.6 · BMI Calculator — Live Build

**Formula:** BMI = weight ÷ height²

Run line by line with the instructor.

```python
# Step 1: Get inputs (float — because decimals are valid)
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in m:  "))

if (weight == 0 and height == 0) :
  print('Enter correct values')

else:
  # Step 2: Calculate
  bmi = weight / (height ** 2)

  # Step 3: Display
  print(f"Your BMI is {bmi:.1f}")
```

```python
# DEMO: why parentheses matter
weight = 70
height = 1.75

correct = weight / (height ** 2)    # BMI = 22.9  ✓
wrong   = (weight / height) ** 2   # BMI = 1600!  ✗

print(f"Correct: {correct:.1f}")
print(f"Wrong:   {wrong:.1f}")
```

### Q2 · Which operator always gives a float, even for whole numbers?

*Try `6 / 2` and `6 // 2` and compare the output.*

```python
# Run to check
print(6 / 2)   # Which type?
print(6 // 2)  # Which type?
print(type(6 / 2))
print(type(6 // 2))
```

<details>
<summary>▶ Show answer</summary>

**Answer: `/` (division)**

`6 / 2` gives `3.0` (a **float**).  
`6 // 2` gives `3` (an **int** — floor division discards the decimal).  

Rule: `/` **always** returns a float. Use `//` when you need a whole-number result.

</details>
