### 1.2 — `range()` Three Forms

````python
# Form 1: range(stop) — starts at 0
print("range(5):")
for i in range(5):
    print(i, end=" ")
print()  # new line
````

````python
# Form 2: range(start, stop) — you choose where to start
print("range(2, 7):")
for i in range(2, 7):
    print(i, end=" ")
print()
````

````python
# Form 3: range(start, stop, step) — skip by a step
print("range(1, 10, 2):")
for i in range(1, 10, 2):
    print(i, end=" ")
print()
````

### 1.3 — Countdown with Negative Step

````python
# Counting backwards — step must be negative
print("Countdown:")
for i in range(5, 0, -1):
    print(i, end=" ")
print("Go!")
````

````python
# See all values at once with list()
print(list(range(3, 8)))       # [3, 4, 5, 6, 7]
print(list(range(0, 10, 3)))   # [0, 3, 6, 9]
````

### Task 1 — Sum of a Range

Ask the user for a **start** number and an **end** number (both integers).
Calculate the **sum of all integers** from start to end (inclusive).
Display the result with an f-string.

**Example:**  
Start: 3, End: 7 → Sum = 3 + 4 + 5 + 6 + 7 = **25**

💡 **Hint:** Remember that `range()` excludes the stop value — you'll need `end + 1`.


````python
# Task 1 — Sum of a Range
# Your code here:
start = int(input("Enter start number: "))
end = int(input("Enter end number"))
sum = 0
for i in range(start,end+1):
  sum+=i
print(sum)
````

### Task 3 — Count Vowels

Ask the user for a **word or sentence**.
Loop through each character and **count how many are vowels** (a, e, i, o, u).
Display the count.

**Example:**  
Input: "Engineering" → Vowels: **5** (E, i, e, e, i)

💡 **Hint:** Use `for ch in text:` and check `if ch.lower() in "aeiou":` to catch both cases.


````python
# Task 3 — Count Vowels
# Your code here:
word = input('Please enter the word: ').lower()
count = 0
for i in word:
  if i == 'a' or i == 'e' or i == 'i' or i=='o' or i == 'u':
    count+=1
print(f'{word}:---> Vowels: {count}')

````
