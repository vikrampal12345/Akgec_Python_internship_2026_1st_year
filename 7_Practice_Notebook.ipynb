### 1.5 — Short-circuit evaluation
Python is lazy — it stops checking as soon as it knows the answer.

````python
# 1.5 — Short-circuit with and
x = 0

# Without the guard, 100/x would crash (ZeroDivisionError)
if x == 0:
  print('Enter a number other than 0')
else:
  if x != 0 or 100/x > 5:
      print("Big ratio")
  else:
      print("x is zero or ratio is small")

# Python sees x != 0 is False → stops immediately
# It never tries 100/0 → no crash!

# Change x to 10 and run again — now both conditions are checked
````
