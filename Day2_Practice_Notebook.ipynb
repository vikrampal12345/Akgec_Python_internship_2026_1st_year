**Q1: What is the difference between `name = "Riya"` and `name == "Riya"`?**


````python
# Click ▶ to see the answer
ans = """
= is ASSIGNMENT — it stores a value under a name.
== is COMPARISON — it asks whether two things are equal (returns True or False).

name = "Riya"    stores the value.
name == "Riya"   asks: is name currently equal to Riya?  Answer: True.
"""
print(ans)
````

**Q2: Why does `int("hello")` crash while `int("95")` works?**


````python
# Click ▶ to see the answer
ans = """
int() only works when the text actually represents a number.
"95" looks like the number 95, so Python can convert it.
"hello" has no numeric meaning — Python raises a ValueError.
"""
print(ans)
````
