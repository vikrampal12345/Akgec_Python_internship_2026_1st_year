### 1.1 · What is a String?

````python
# Strings can use single or double quotes — both work
name  = "Priya"
city  = 'Ghaziabad'
empty = ""

print(name)
print(type(name))   # <class 'str'>

# len() tells you the number of characters
print(len(name))    # 5
print(len(city))    # 10
print(len(empty))   # 0
````

### 1.2 · Concatenation and Repetition

````python
# + joins strings together
first = "Hello"
second = "World"
print(first +" "+ second)

````

````python
print(first*5)
````

````python
# * repeats a string
print("Ha" * 3)       # HaHaHa
print("=" * 20)       # ====================
print("-" * len(name))  # -----  (same length as name)
````

### 1.3 · f-strings: Deeper

````python
# Multiple variables in one f-string
name  = "Riya"
score = 92
print(f"{name} scored {score}/100")

# Expressions inside the braces
a, b,c,d = "Lord","of","the",'Rings'
print(f"{a} + {b} +{c}+{d} = {a*b*c*d}")
print(a*b*c*d)

# Format specifiers (you saw :.1f in the BMI calculator)
pi = 3.14159
print(f"pi ≈ {pi:.2f}")    # 2 decimal places
print(f"pi ≈ {pi:.4f}")    # 4 decimal places
````

### 1.4 · String Indexing — Positive Indices

Strings are sequences. Every character has a position (called an **index**), starting at **0**.

```
s = "Python"
     P  y  t  h  o  n
     0  1  2  3  4  5   ← positive indices
```

````python
# Deliberately trigger an IndexError — run it!
# s[6]  would crash — "Python" only has indices 0 to 5

# Read the error message: it tells you exactly what went wrong
print(s[6])   # Uncomment to see the error
````

### 1.5 · Negative Indices — Count from the Right

```
s = "Python"
     P   y   t   h   o   n
    -6  -5  -4  -3  -2  -1   ← negative indices
```

`s[-1]` is **always** the last character — no matter how long the string is.

````python
s = "Python"

print(s[-1])   # Last character
print(s[-2])   # Second-to-last
print(s[-6])   # Same as s[0]

# Useful: last character of ANY string
word = input("Enter a word: ")
print(f"First: {word[0]}")
print(f"Last:  {word[-1]}")
````

### 1.6 · String Slicing  `[start : stop : step]`

> **Key rule: the stop index is EXCLUDED.**  
> `s[0:3]` gives indices 0, 1, 2 — that is 3 characters, not 4.

Predict the output of each line **before** running.

````python
[start,stop)
````

````python
s = "Python"
s[1:4:2]
````

````python
x = 'Palantir'
x[:-1:]
x[:7]
````

````python
y = "Mordor"
x[::-1]
````

````python
s = "Python"

# Basic slicing
print(s[0:3])    # From index 0 up to (not including) 3
print(s[3:6])    # From index 3 up to (not including) 6

# Omitting start or stop
print(s[:3])     # From beginning to index 3 (excluded)
print(s[3:])     # From index 3 to the end
print(s[:])      # Entire string — a copy
````

````python
s = "Python"

# Using step
print(s[::2])    # Every 2nd character: P, t, o
print(s[1::2])   # Every 2nd starting from index 1: y, h, n

# The magic trick: step of -1 walks backwards → reverses the string!
print(s[::-1])   # Reversed

# Try it with your own name:
name = input("Enter your name: ")
print(f"{name} reversed is {name[::-1]}")
````

### Q2 · What does `"Python"[0:3]` return?
*Remember the stop-is-excluded rule.*

````python
print("Python"[0:3])
print(len("Python"[0:3]))  # How many characters?
````

<details>
<summary>▶ Show answer</summary>

**Answer: `'Pyt'` — 3 characters, not 4**

The slice `[0:3]` gives indices 0, 1, 2. Stop index 3 is **excluded**.

### Q3 · What does `"Hello"[::-1]` return?

````python
print("Hello"[::-1])
````

<details>
<summary>▶ Show answer</summary>

`[::-1]` is the standard Python idiom for reversing a string — no loops needed.
