### 1.1 — Pattern 1: Accumulator

**Question it answers:** *What's the total?*

Template:
```
total = 0          # start value
for item in data:
    total += item  # accumulate
print(total)       # result
```


````python
# Accumulator: sum of even numbers from 1 to 50
total = 0
for i in range(2, 51, 2):
    total += i
print(f"Sum of even numbers 1–50 = {total}")
````

````python
# Accumulator: product (multiply instead of add)
product = 1  # start at 1 for multiplication, not 0!
for i in range(1, 6):
    product *= i
print(f"5! (factorial) = {product}")
````

### 1.2 — Pattern 2: Counter

**Question it answers:** *How many match?*

Template:
```
count = 0
for item in data:
    if condition:
        count += 1     # match found
print(count)
```


````python
# Counter: how many numbers from 1–100 are divisible by both 3 AND 5?
count = 0
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        count += 1
print(f"Numbers divisible by both 3 and 5: {count}")
````

````python
# Counter: count uppercase letters in a sentence
sentence = "Hello World, Python is Great!"
upper_count = 0
for ch in sentence:
    if ch.isupper():
        upper_count += 1
print(f"Uppercase letters: {upper_count}")
````

### 1.3 — Pattern 3: Search

**Question it answers:** *Is there one that...?*

Template:
```
found = False
for item in data:
    if condition:
        found = True
        break          # stop early
print(found)
```


````python
# Search: is there a negative number in the list?
nums = [4, 7, -2, 9, 1]
found = False
for n in nums:
    if n < 0:
        found = True
        break
print(f"Negative found: {found}")
````

````python
# Search variation: find and report the FIRST match
nums = [10, 25, 3, 42, 7, 18]

first_even = None
for n in nums:
    if n % 2 == 0:
        first_even = n
        break

if first_even is not None:
    print(f"First even number: {first_even}")
else:
    print("No even numbers found")
````

### 1.4 — Prime Checker (Capstone)

The prime checker combines the **search pattern** with maths.
A prime number has no divisors other than 1 and itself.


````python
# Prime checker — search for a divisor
number = int(input("Write a number to check: "))
if number > 1:
  is_prime = True
  for i in range(2,number):
      if number % i == 0:
        is_prime = False
        break
  print(f'Is your number prime? {is_prime}')
else:
  print('Enter a number greater than 1')
````
