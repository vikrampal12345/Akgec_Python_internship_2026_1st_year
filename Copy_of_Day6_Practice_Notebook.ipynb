### 1.3 — `elif`: Multiple branches (Grade Classifier)
When you need more than two options, chain `elif` between `if` and `else`.

````python
# 1.3 — Grade classifier with elif
# Python checks top-to-bottom and STOPS at the first True condition

# >=90 --A
# 80 -90 B
#50 - 80 C
#<50 D

score = int(input("Write the student's score: "))
if score >= 90:
  print("The grade is A")
elif score >=80 and score <90:
  print("The grade is B")
elif score>=50 and score <80:
  print("The grade is C")
else:
  print("The score is D")
````

````python
# 1.3b — Try different scores to see each branch fire
# Change score to: 95, 72, 60, 45 — predict the grade BEFORE running

score = 72

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
````

````python
# 1.3c — WHY ORDER MATTERS
# Watch what happens when we check the LOWEST threshold first (WRONG!)

score = 95

if score >= 60:       # ← 95 >= 60 is True — Python stops here!
    grade = "D"
elif score >= 70:
    grade = "C"
elif score >= 80:
    grade = "B"
elif score >= 90:
    grade = "A"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
print("Oops — a 95 got a D! Always check the highest threshold first.")
````

**Q1.** What happens if you forget the colon `:` after an `if` condition?

<details>
<summary>Reveal answer</summary>

You get a `SyntaxError`. The colon tells Python that the indented body starts on the next line. Without it, Python doesn't know where the condition ends.

</details>

---

**Q2.** In a grade classifier, why must you check `>= 90` before `>= 60`?

<details>
<summary>Reveal answer</summary>

Because Python checks conditions top-to-bottom and **stops at the first True**. If you check `>= 60` first, a score of 95 would match that condition and get the wrong grade (D instead of A). Always check the highest threshold first.

</details>

---

**Q3.** What is the difference between `if/elif/else` and three separate `if` statements?

<details>
<summary>Reveal answer</summary>

With `if/elif/else`, **exactly one branch runs** — Python stops at the first True condition.

With three separate `if` statements, **all three are checked independently** — multiple branches could run. Use `if/elif/else` for mutually exclusive categories (like grades), and separate `if` statements for independent checks.

```python
# Exactly one branch runs:
if x > 10:
    print("big")
elif x > 5:
    print("medium")
else:
    print("small")

# Both could run:
if x > 5:
    print("above 5")
if x > 3:
    print("above 3")
```

</details>
