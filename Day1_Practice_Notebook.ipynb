### 1.1 — Your first command: `print()`
`print()` puts whatever you give it onto the screen. Text goes **inside quotes**.

````python
print("Hello, world!")
````

Now change the words and run again. The output changes to match.

````python
print("I am learning Python at AKGEC")
````

### 1.2 — Printing more than one thing
Separate items with a **comma**. Python prints them on one line with a space between.

````python
print('Hello first-year engineers')
````

### 1.3 — Asking a question: `input()`
`input()` shows a question, **pauses**, waits for the user to type, and hands back whatever they typed.
We store that answer in a **label** called `name` so we can use it on the next line.

````python
name = input("What is your name? ")
age = input("What is your age?")
print("Hello,", name)
print("Age of", name, "is",age)
````

Run the cell above. A box appears asking for your name — type it, press **Enter**, and watch the greeting.

> `name` is just a label that holds the answer. We'll learn much more about labels (variables) tomorrow.

### 1.4 — Python figures out the type for you
Every value has a **type**. You never tell Python the type — it works it out from the value.
`type(...)` shows you what Python decided.

````python
#print(type("My name is Amit"))   # text  -> str
#print(type(18))       # whole number -> int
print(type(5.5))      # decimal -> float
print(type(True))     # yes/no -> bool
````

````python
str(34)
````

````python
bool(0)
````

That automatic behaviour is called **dynamic typing**. For today, just notice: Python never *asked* you the type.
